```python
import pandas as pd
import numpy as np
from scipy.spatial.distance import euclidean, cityblock, minkowski, hamming
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('abalone.csv',header=None)
df.head()
```

```python
df.columns=['Sex','Length','Diameter','Height','Whole_weight','Shucked_weight','Viscera_weigh','Shell_weight','Rings']
```

```python
df.head()
```

```python
# Encode categorical column ('Sex') to numerical values
label_encoder = LabelEncoder()
df['Sex'] = label_encoder.fit_transform(df['Sex'])  # Converts 'M', 'F', 'I' to numbers
```

```python
# Select two sample points (rows 0 and 1)
point1 = df.iloc[0].drop("Rings").values  # Exclude target variable
point2 = df.iloc[1].drop("Rings").values
```

```python
### **1️⃣ Euclidean Distance (Manual)**
def euclidean_distance(p1, p2):
    return sum((p1 - p2) ** 2) ** 0.5

### **2️⃣ Manhattan Distance (Manual)**
def manhattan_distance(p1, p2):
    return sum(abs(p1 - p2))

### **3️⃣ Minkowski Distance (Manual, p=3)**
def minkowski_distance(p1, p2, p):
    return sum(abs(p1 - p2) ** p) ** (1/p)

### **4️⃣ Hamming Distance (Manual)**
def hamming_distance(p1, p2):
    return sum(p1 != p2) / len(p1)

# Compute Distances
euclidean_dist = euclidean_distance(point1, point2)
manhattan_dist = manhattan_distance(point1, point2)
minkowski_dist = minkowski_distance(point1, point2, p=3)
hamming_dist = hamming_distance(point1, point2)

# Print Results
print(f"Euclidean Distance: {euclidean_dist}")
print(f"Manhattan Distance: {manhattan_dist}")
print(f"Minkowski Distance (p=3): {minkowski_dist}")
print(f"Hamming Distance: {hamming_dist}")
```
